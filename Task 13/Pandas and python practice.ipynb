# Update coulmn values in Marital Status to a charcheter crossponding it's value
for example Single ->S ,Married->M and so on

```python
alumni['Marital Status']=alumni['Marital Status'].replace({
    'Single': 'S',
    'Married': 'M',
    'Divorced': 'D',
    'Widowed': 'W',
})
print(alumni.columns)
```

# Drop the Fee coulmn

```python
alumni.drop(columns=['Fee'], inplace=True, errors='ignore')
```
